# how to deal with excel

```python
# imporing excel file in
df=pd.read_excel('sampledatainsurance.xlsx',sheet_name='PolicyData')
```

```python
df.head(5)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#pivot table
# by default aggfun is mean in pivot table
df1=df.pivot_table(values='InsuredValue',index='BusinessType',columns='Location',aggfunc=np.sum)
```

```python
#fillimg nan values with 0
df2=df1.fillna(value=0)
```

```python
#export excel file
df2.to_excel('Test sample insurance data.xlsx',sheet_name='output')
```

# Dealing with 2 million records

```python
#importing big data of 2m sales records
df=pd.read_csv('2m Sales Records.csv')
```

```python
df['Country'].value_counts()
```

```python
#i am interested only in total revenue and total profit
df1=df.groupby('Country').sum()[['Total Revenue','Total Profit']]/1000000
df1
```
