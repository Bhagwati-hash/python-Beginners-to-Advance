```python
# creting a time line that i want to analyze the data of above global banks from year 2006 to year 2016
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
# citi group
C=data.DataReader('C','yahoo',start,end)
```

```python
#Goldmansachs , jpmorgan chase ,Morgan stanley
GS=data.DataReader('GS','yahoo',start,end)
JPM=data.DataReader('JPM','yahoo',start,end)
MS=data.DataReader('MS','yahoo',start,end)
```

```python
MS.head()
```

```python
tickers=['BAC','C','GS','JPM','MS']
```

```python
# CONCATENATE
bank_stocks=pd.concat([BAC,C,GS,JPM,MS],axis=1,keys=tickers)
#axis=1 will concatenate according to columns
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names=['Bank,Ticker','Stock info']
```

```python
bank_stocks.head()
```

# problem statements



**question 1 : what is the maximum close price for each back's stock throughout the time period**


```python
bank_stocks['BAC']['Close'].max()
```

```python
# ONE RULE in Programming is DRY ' do not rpeat yourself'
# so we will create a dÂ¨for loop
```

```python
tickers
```

```python
# for loop for getting max clsoe price for all of the 5 banks at once
for i in tickers:
    print('The answer of',i,'is',bank_stocks[i]['Close'].max())
```

```python
# cross section
# cross section is used for multilevel indexing
bank_stocks.head()
```

```python
# we can extract max close price for all the banks using crossection
bank_stocks.xs(key='Close',axis=1,level='Stock info').max()
```

**returns = (value at time t - value at time (t-1))/ value at time (t-1)**

```python
returns=pd.DataFrame()
```

```python
# for loop for getting return percentage for all of the banks
for i in tickers:
    returns[i  +  'return']=bank_stocks[i]['Close'].pct_change()
```

```python
returns
```

**question: 2 which dates each banks has the best and worst single day returns**

```python
# the worst date for each of the bank
returns.idxmin()
```

```python
# the best date for each of the bank
returns.idxmax()
```

**question:3**

   **b.Which whould you classify as the riskiest during 2015**
