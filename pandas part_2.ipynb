# handling missing values

```python
import pandas as pd
import numpy as np
```

```python
my_dict={'india':[20,30,np.nan],'usa':[20,np.nan,np.nan],'china':[60,20,60]}
```

```python
my_dict
```

```python
df=pd.DataFrame(data=my_dict,index=['state1','state2','state3'])
```

```python
df
```

```python
#how to drop rows with null values
df.dropna(axis=0)
```

```python
#how to drop columns with null vales
df.dropna(axis=1)
```

```python
#threshhold 
df.dropna(thresh=2)
#atleast two values in arow which are not null this is what means by thtresh 2 thats why it is showing the following reasult
```

```python
df.dropna(axis=1,thresh=2)
```

```python
df
```

```python
#filling missing values with some value
df.fillna(value=0)
```

```python
#suppose i want to fill the nan values of column india with mean of that column
df['india'].fillna(value=df['india'].mean())
```

```python
df['usa'].fillna(value=df['usa'].max())
```

# concatinating , merging and joining

```python
# concatinating (glues the data frame together according to rows or column)
#concat
my_data1={'india':[1,2,3,4],'brazil':[5,6,7,8],'usa':[9,10,11,12],'china':[13,14,15,16]}
```

```python
df1=pd.DataFrame(data=my_data1,index=['A','B','C','D'])
```

```python
df1
```

```python
my_data2={'india':[10,20,30,40],'brazil':[50,60,70,80],'usa':[90,100,110,120],'china':[130,140,150,160]}
```

```python
df2=pd.DataFrame(data=my_data2,index=['E','F','G','H'])
```

```python
df2
```

```python
my_data3={'india':[100,200,300,400],'brazil':[500,600,700,800],'usa':[900,1000,1100,1200],'china':[1300,1400,1500,1600]}
```

```python
df3=pd.DataFrame(data=my_data3,index=['P','Q','R','S'])
```

```python
df3
```

```python
#by default concat command works according to row
df4=pd.concat([df1,df2,df3])
```

```python
df4
```

```python
#axis =1 according to column
df5=pd.concat([df1,df2,df3],axis=1)
```

```python
df5
```

```python
# merging
df4=pd.DataFrame(data={'country':['india','uk','usa','denmark'],
                      'north':[100,200,300,400],
                      'south':[500,600,700,800]})
```

```python
df5=pd.DataFrame(data={'country':['india','uk','usa','denmark'],
                      'east':[200,400,100,600],
                      'west':[500,800,500,900]})
```

```python
df=pd.merge(left=df4,right=df5,how='inner',on='country')
```

```python
df
```

```python
#merging based on 2 keys rather one key
df6=pd.DataFrame(data={'key1':['k0','k1','k0','k3'],'key2':['k4','k5','k6','k7'],
                       'A':['A0','A1','A2','A3'],'B':['A4','A5','A6','A7']})
df7=pd.DataFrame(data={'key1':['k0','k1','k0','k4'],'key2':['k4','k5','k6','k4'],
                       'C':['C0','C1','C2','C3'],'D':['D4','D5','D6','D7']})
```

```python
df6
```

```python
df7
```

```python
pd.merge(left=df6,right=df7,how='inner',on=['key1','key2'])
#it will show connect rows which are common in df6 and df7
```

```python
#outer joint will show all values
pd.merge(left=df6,right=df7,how='outer',on=['key1','key2'])
```

```python
#left and right joint
pd.merge(left=df6,right=df7,how='left',on=['key1','key2'])
```

```python
pd.merge(left=df6,right=df7,how='right',on=['key1','key2'])
```

```python
#cross joint
school=pd.DataFrame(data=['dav','apj','convent','govt'],columns=['school name'])
```

```python
school
```

```python
kids=pd.DataFrame(data=['tejas','bhagwati'],columns=['kids name'])
```

```python
kids
```

```python
pd.merge(left=school,right=kids,how='cross')
```

```python
#join function 
#dealing with rows
df8=pd.DataFrame(data={'A':['A0','A1','A2'],'B':['B0','B1','B2']},index=['k0','k1','k2'])
df9=pd.DataFrame(data={'C':['c0','c2','c3'],'D':['d0','d2','d3']},index=['k0','k2','k3'])
```

```python
df8
```

```python
df9
```

```python
#inner join by default it take axis=1
df8.join(df9,how='inner')
#we could see thoses key values which are present in both the data frames
```

```python
df8.join(df9,how='right')
```
